```python
import pandas as pd
df = pd.read_excel('MorteSuspeita_2013_2022.xlsx')
df
```

## Remoção de dados duplicados

```python
# Remover duplicado
df = df.drop_duplicates()
# Transformar para toLower
df = df.applymap(lambda s: s.lower() if type(s) == str else s)
#Removendo espaços inuteis
df = df.applymap(lambda s: s.strip() if type(s) == str else s)
df
```

## Filtragem de dados da cidade de SP

```python
filtro = df['MUNICIPIO_CIRCUNSCRICAO'] == 's.paulo'
df = df.loc[filtro, :]
#df2 = df[filtro]
#df2 = df.query["MUNICIPIO_CIRCUNSCRICAO == 's.paulo'"]
df
```

## Transformar string para minúsculo e padronizar


```python
import unidecode
# Transformar para toLower
df = df.applymap(lambda s: s.lower() if type(s) == str else s)
# Através de uma pesquisa descobri a biblioteca unidecode que permite fazer essas operações
df = df.applymap(lambda s: unidecode.unidecode(s) if type(s) == str  else s)
df
```

## Transformar dados temporais para o tipo apropriado

```python
# ver o tipo dos dados atuais temporais
df[["DATAHORA_REGISTRO_BO"]].dtypes
```

```python
import string
df[["DATAHORA_REGISTRO_BO"]] =  df[["DATAHORA_REGISTRO_BO"]].astype('string')

df[["DATAHORA_REGISTRO_BO"]].dtypes
```
